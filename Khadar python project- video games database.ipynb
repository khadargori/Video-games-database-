```python
#16. remove all null values from User_Score and print those values to check
print("16. user_score values after removing null values:")
video_games.dropna(subset=['User_Score'],inplace=True)
video_games['User_Score']
```

# 17. plot the histogram chart with x=User_Score and y-Developer 
import seaborn as sns
print("17. Histogram of User_Score (grouped by developer:")
top_developers=video_games['Developer'].value_counts().nlargest(5).index
filtered_games=video_games[video_games['Developer'].isin(top_developers)]
mplt.figure(figsize=(12,8))
sns.histplot(data=filtered_games,x='User_Score',hue='Developer',kde=True)
mplt.title('Distribution of user scores by top 5 developers')
mplt.xlabel('User Score')
mplt.ylabel('Developer')
mplt.show()

```python
# 18. Display all video games.
print("all videol games:")
print(video_games['Name'].unique())
```

```python
# 19. update NA_SALES with +5: (mapwith lambda)
#showing visual is extra.
print("19. NA_Sales afer adding 5:")
video_games['NA_Sales']=video_games['NA_Sales'].map(lambda x: x + 5 if pd.notna(x) else x)
print(video_games['NA_Sales'])
mplt.figure(figsize=(12,6))
sns.histplot(video_games['NA_Sales'])
mplt.title('Distribution of NA_Sales')
mplt.show()
```

```python
#20. Display which Genre has many games.
print("20. Number of games per Genre:")
all_games=video_games.groupby('Genre')['Genre'].count()
print(all_games)
```

```python
#21. Create a bar char for all games (Genre Distribution)
print("21. Bar chart of games by Genre:")
mplt.figure(figsize=(12,6))
all_games.plot(kind='bar')
mplt.title("Number of Games per Genre:")
mplt.xlabel('Genre')
mplt.ylabel('Number of games')
mplt.show()
```

```python
#22. which developer has most games sold: top 20 developers 
print(" 22. top 20 most sold games by developers:\n")
top_20_developers=video_games.groupby('Developer')['Global_Sales'].sum().nlargest(20)
print(top_20_developers)
```

```python
#23. plot the bar chart for the above top 20 developers
print("23. bar chart for top 20 developers:")
mplt.figure(figsize=(14,8))
top_20_developers.plot(kind='bar')
mplt.title('top 20 developers by Global sales')
mplt.xlabel('Developer')
mplt.ylabel('Global Sales')
mplt.show()
```
